# ダミーデータの作成
## 【問題1】乱数の作成

```python
# モジュールのインポート
import numpy as np

# 平均値及び共分散の設定
mean = (-3, 0)
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

# seedを固定し乱数を作成
np.random.seed(0)
data = np.random.multivariate_normal(mean, cov, size=500)

# 値の確認
print(data.shape)
print(np.mean(data, axis=0))
print(np.cov(data, rowvar=False))
```

## 【問題2】散布図による可視化

```python
# モジュールのインポート
import matplotlib.pyplot as plt

# 散布図の作成
plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(data[:, 0], data[:, 1]);
```

## 【問題3】ヒストグラムによる可視化

```python
# ヒストグラムの作成
plt.title('histogram of x1')
plt.xlabel('x1')
plt.ylabel('frequency')
# オプションとしてbinsを50(defoult=20)に、範囲を下記グラフと揃えるため設定
plt.hist(data[:, 0], bins=50, range=(-6, 3));
```

```python
plt.title('histogram of x2')
plt.xlabel('x1')
plt.ylabel('frequency')
plt.hist(data[:, 1], bins=50, range=(-6, 3));
```

## 【問題4】データの追加

```python
# 平均値及び共分散の設定を行い乱数の作成
mean_2 = (0, -3)
data_2 = np.random.multivariate_normal(mean_2, cov, size=500)

# data_2の登場によってわかりづらくなったのでdataをrename
data_1 = data
```

```python
# 散布図の作成
plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(data_1[:, 0], data_1[:, 1], label=0)
plt.scatter(data_2[:, 0], data_2[:, 1], label=1)
plt.legend();
```

## 【問題5】データの結合

```python
#　データを縦方向に結合
data_con = np.concatenate([data_1, data_2], axis=0)

print(data_con.shape)
```

## 【問題6】ラベル付け

```python
# 全てが0、1の値をとる(500, 1)のラベル付用の配列を作成
data_0_full = np.full((500, 1), 0)
data_1_full = np.full((500, 1), 1)

# 書くデータに横方向に結合(ラベル付)
data_1_label = np.concatenate([data_1, data_0_full], axis=1)
data_2_label = np.concatenate([data_2, data_1_full], axis=1)

# ラベル付されたデータを縦に結合
data_con_label = np.concatenate([data_1_label, data_2_label], axis=0)

print(data_con_label)
```
